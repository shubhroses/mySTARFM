```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# Create a grid of coordinates
x, y = np.meshgrid(np.arange(30), np.arange(30))

# Calculate the distance from the center of the circle
distance = np.sqrt((x - 15)**2 + (y - 15)**2)

# Create a boolean mask for the circle
circle_mask = np.logical_and(distance > 13, distance < 15)

# Create the final array with the circle
circle_array = np.zeros((30, 30))
circle_array[circle_mask] = 1
```

```python
# create an array of zeros with shape (30, 30)
square_array = np.zeros((30, 30))

# draw a diagonal line from (0, 0) to (29, 29) with thickness 3
for i in range(3):
    diagonal = np.ones(30 - i*2, dtype=bool)
    square_array[np.diag(diagonal, k=i)] = 1
    square_array[np.diag(diagonal, k=-i)] = 1

```

```python
plt.imshow(circle_array, cmap='gray')
plt.show()
```

```python
# convert the array to PIL Image
original_image = Image.fromarray(np.uint8(circle_array * 255), mode='L')

# resize the image to shape (100, 100) using nearest neighbor interpolation
resized_image = original_image.resize((100, 100), resample=Image.NEAREST)

# convert the resized image back to numpy array
new_array = np.array(resized_image) / 255
```

```python
plt.imshow(new_array, cmap='gray')
plt.show()
```
